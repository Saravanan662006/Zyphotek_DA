```python
df = pd.read_excel("Zyphotek.xlsx")
df.head()
```

```python
print(df.info())
print(df.describe(include="all"))
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
df.fillna("-", inplace=True)
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df.to_excel("cleaned_college_data.xlsx", index=False)
```

```python
state_count = df["Location"].value_counts()
print(state_count)
```

```python
naac = df["NAAC_Grade"].value_counts()
print(naac)
```

```python
state_count.plot(kind="bar", figsize=(10,5))
plt.title("Colleges by State")
plt.xlabel("State")
plt.ylabel("Number of Colleges")
plt.show()
```

```python
naac.plot(kind="pie", autopct="%1.1f%%")
plt.title("NAAC Grade Distribution")
plt.ylabel("")
plt.show()
```
